# DATA ENGINEERING
1. Read the data in datewise fashion
2. Split into morning, afternoon and evening datasets for more granular processing

```python
def get_date(file_name):
    date_pattern = r'\d{8}'  # Matches 8 digits (YYYYMMDD)
    match = re.search(date_pattern, file_name)
    if match:
        date_str = match.group()
        return date_str
    else:
        return np.nan

def get_files(start_date, end_date):
    filename_list = []

    while start_date <= end_date:
        filename = f'QR_TAKEHOME_{str(start_date.strftime("%Y%m%d"))}.csv.parquet'
        filename_list.append(filename)
        start_date += timedelta(days=1)

    return filename_list

directory = '/Projects/ClearStreet/qr_takehome'

morning_df = pd.DataFrame()
midday_df = pd.DataFrame()
afternoon_df = pd.DataFrame()

start_date = datetime(2022, 2, 1)
end_date = datetime(2022, 2, 28)
filename_list = get_files(start_date, end_date)

for filename in filename_list:
    filepath = os.path.join(directory, filename)
    if not os.path.exists(filepath):
        continue

    parquet_table = pq.read_table(filepath)
    df_datewise = parquet_table.to_pandas()
    df_datewise['date'] = pd.to_datetime(get_date(filename), format='%Y%m%d')

    morning_df = pd.concat([morning_df, df_datewise.loc[df_datewise['time'] <= 39600000].sort_values(by='time')], ignore_index=True)
    midday_df = pd.concat([midday_df, df_datewise.loc[(df_datewise['time'] >= 39601000) & (df_datewise['time'] <= 50400000)].sort_values(by='time')], ignore_index=True)
    afternoon_df = pd.concat([afternoon_df, df_datewise.loc[(df_datewise['time'] >= 50401000)].sort_values(by='time')], ignore_index=True)

data = morning_df.copy()
df = morning_df
df['time'] = pd.to_timedelta(df['time'], unit='ms')
df['datetime'] = df['date'] + df['time']
df['datetime'] = pd.to_datetime(df['datetime'])
df
```

## Handling NaNs
- Drop records where Q1 or Q2 < 1
- As said replace records with values 999999 as NaN
- Drop records where more than 60% of values are NaN

```python
df = data.loc[data['Q1']>=0.99].loc[data['Q2']>=0.99]
df.replace(999999, np.nan, inplace=True)
df = df.dropna(thresh=int(0.5 * df.shape[1]))
df.reset_index(inplace=True, drop=True)
df
```

#### Even using PCA feature selection cannot be boiled down to <10 dimensions. That is still a lot of dimensionality in the data, we need to find some other techniques to reduce dimensionality. Lets see some possible techniques to try here
1. Only sticking to PCA
    - Reduce the 'n_components' variable even further. This parameter encapsulates the dataset's explained variance threshold.
    - Calibrate the threshold to a level where there are max 10 dimensions after applying PCA
    - Apply a weighted OLS or simple OLS and see how the model performs
2. Combining PCA with regularization
    - Use LASSO on the 99% explained variance PCs
3. Use modified LASSO
    - We have seen earlier that WLS performs well with multiregession but OLS is not as great
    - The penalty term would be modified based on the WLS calculation of residual and not OLS

```python
X = sm.add_constant(pc_df[[f'PC{i}' for i in range(pca.n_components_)]])
Y = pc_df['Y']

ols_model = sm.OLS(Y, X).fit()
weights = 1 / abs(ols_model.resid)
model = sm.WLS(Y, X, weights=weights).fit()

bp_test = het_breuschpagan(model.resid, model.model.exog)
white_test = het_white(model.resid, model.model.exog)
labels = ['Lagrange Multiplier statistic', 'p-value', 'f-value', 'f p-value']
print(dict(zip(labels, bp_test)))
print(dict(zip(labels, white_test)))
model.summary()
```

The PCA regression is clearly not giving us good results. $ R^2 $ is clearly low, and there are extremely strong indications of heteroskedasticity. We should try out 2nd method using PCA combined with LASSO to completely eliminate some features if needed

### 2. Using PCA with LASSO

```python
X = df[[f'X{i}' for i in range(1, 376)]]
Y1 = pc_df['Y']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
pca = PCA(n_components=0.99)
X1_pca = pca.fit_transform(X_scaled)
#explained_variance = pca.explained_variance_ratio_
#print(f"Explained variance: {explained_variance}")
plt.figure(figsize=(10, 6))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
plt.show()
```

```python
pc_df = pd.DataFrame(data=X1_pca, columns=[f'PC{i}' for i in range(pca.n_components_)])
pc_df['Y'] = Y1
X = pc_df[[f'PC{i}' for i in range(pca.n_components_)]]
scaler = StandardScaler()
X = scaler.fit_transform(X)
Y = pc_df['Y']
alpha = 0.1
lasso = Lasso(alpha=alpha)
lasso.fit(X, Y)
y_pred = lasso.predict(X)
r_squared = r2_score(Y, y_pred)
print(f"R-squared: {r_squared}")
coefficients = pd.DataFrame({'Feature': [f'PC{i}' for i in range(pca.n_components_)], 'Coefficient': lasso.coef_})
coefficients.loc[coefficients['Coefficient'] != 0]
```

PCA with LASSO is also underperforming compared to our complex overfitted WLS model, and is in fact has worst performance so far. The penalty is gargantuan for all coefficients and LASSO says that none of the PCs are relevant for higher levels of penalty

### 3. Modified LASSO

```python
X = df[[f'X{i}' for i in range(1, 376)]]
scaler = StandardScaler()
X = scaler.fit_transform(X)
Y = df['Y1'].values.reshape(-1, 1)
Y = scaler.fit_transform(Y)

ols_model = sm.OLS(Y, sm.add_constant(X)).fit()
weights = 1 / abs(ols_model.resid)

def weighted_lasso(X, y, weights, alpha=1.0):
    """
    Fits a weighted LASSO model.
    """
    # Scale X and y by weights
#     W = np.sqrt(weights)
#     X_w = np.array(X) * np.array(W)[:, np.newaxis]
#     y_w = np.array(y) * W

    # Fit LASSO
    lasso = Lasso(alpha=alpha)
    lasso.fit(X, y)
    return lasso

lasso = weighted_lasso(X, Y, weights, alpha=0.1)
y_pred = lasso.predict(X) # * np.sqrt(weights)[:, np.newaxis])

r_squared = r2_score(Y, y_pred)
print(f"R-squared: {r_squared}")
coefficients = pd.DataFrame({'Feature': [f'X{i}' for i in range(1, 376)], 'Coefficient': lasso.coef_})
coefficients.loc[coefficients['Coefficient'] != 0]
```

```python
#### Testing the model on unknown data
def get_date(file_name):
    date_pattern = r'\d{8}'  # Matches 8 digits (YYYYMMDD)
    match = re.search(date_pattern, file_name)
    if match:
        date_str = match.group()
        return date_str
    else:
        return np.nan

def get_files(start_date, end_date):
    filename_list = []

    while start_date <= end_date:
        filename = f'QR_TAKEHOME_{str(start_date.strftime("%Y%m%d"))}.csv.parquet'
        filename_list.append(filename)
        start_date += timedelta(days=1)

    return filename_list

def get_data(start_date, end_date):
    directory = '/Projects/ClearStreet/qr_takehome'

    morning_df = pd.DataFrame()
    filename_list = get_files(start_date, end_date)

    for filename in filename_list:
        filepath = os.path.join(directory, filename)
        if not os.path.exists(filepath):
            continue

        parquet_table = pq.read_table(filepath)
        df_datewise = parquet_table.to_pandas()
        df_datewise['date'] = pd.to_datetime(get_date(filename), format='%Y%m%d')

        morning_df = pd.concat([morning_df, df_datewise.loc[df_datewise['time'] <= 39600000].sort_values(by='time')], ignore_index=True)

    df = morning_df
    df['time'] = pd.to_timedelta(df['time'], unit='ms')
    df['datetime'] = df['date'] + df['time']
    df['datetime'] = pd.to_datetime(df['datetime'])

    return df

def handle_nan(data):
    df = data.loc[data['Q1']>=0.99].loc[data['Q2']>=0.99]
    df.replace(999999, np.nan, inplace=True)
    df = df.dropna(thresh=int(0.5 * df.shape[1]))
    df.reset_index(inplace=True, drop=True)

    numeric_columns = df.select_dtypes(include=['number']).columns
    columns_to_interpolate = [col for col in numeric_columns if col not in ['date', 'time', 'datetime']]
    df[columns_to_interpolate] = df[columns_to_interpolate].interpolate(method='linear')

    percentile_low = 0.001
    percentile_high = 0.999
    numeric_columns = ['Y1']
    for col in numeric_columns:
        max_val = df.groupby('date')[col].transform(lambda x: x.quantile(percentile_high))
        min_val = df.groupby('date')[col].transform(lambda x: x.quantile(percentile_low))
        df[col] = np.where((df[col] > min_val) & (df[col] < max_val), df[col], np.nan)
    df = df.dropna().reset_index(drop=True)

    return df

def predict(model, X, Y, model_type='default'):
    if model_type == 'log_regression':
        min_values = X.min()
        X = X - min_values
        X = np.log(1 + X)  # Independent variable
        X = sm.add_constant(X)
        Y = np.log(1 + (Y - Y.min()))

    predictions = model.predict(sm.add_constant(X))
    return Y, predictions

def out_of_sample_test(model, top_features):
    start_date = datetime(2022, 3, 1)
    end_date = datetime(2022, 3, 6)
    df = get_data(start_date, end_date)
    df = handle_nan(df)

    X = scaler.fit_transform(df[top_features])
    Y1 = scaler.fit_transform(df['Y1'].values.reshape(-1, 1)) #df['Y1']
    Y1, predictions = predict(model, X,Y1)

    mse = mean_squared_error(Y1, predictions)
    r2 = r2_score(Y1, predictions)

    print(f"Mean Squared Error: {mse}")
    print(f"R-squared: {r2}")

    plt.figure(figsize=(15, 8))
    plt.plot(predictions, color='maroon', label="Predictions")
    plt.plot(Y1, color='blue', label="Actual")
    plt.xlabel('Time')
    plt.ylabel('Value')
    plt.legend()
    plt.show()
```

### Feature extraction using LSTM

```python
X = df[[f'X{i}' for i in range(1, 376)]].copy()
X['Y1_lag'] = df['Y1'].shift(1)
X.dropna(inplace=True)
Y = df['Y1'][1:]
scaler = StandardScaler()
X = scaler.fit_transform(X)
window_size = 10  # Adjust the window size as needed
step = 10  # Adjust the step size as needed
sequences = []
target_values = []

for i in range(0, len(X) - window_size + 1, step):
    sequences.append(X[i:i+window_size])
    target_values.append(Y[i+window_size-1])

sequences = np.array(sequences)
target_values = np.array(target_values)
split_index = int(len(sequences) * 0.8)
X_train, X_test = sequences[:split_index], sequences[split_index:]
Y_train, Y_test = target_values[:split_index], target_values[split_index:]
input_dim = X_train.shape[2]  
encoding_dim = 8 

# Define the encoder
encoder_input = keras.Input(shape=(window_size, input_dim))
encoded = keras.layers.LSTM(encoding_dim, activation='tanh', return_sequences=True)(encoder_input)
encoded = keras.layers.LSTM(encoding_dim, activation='tanh')(encoded)
encoder_model = keras.Model(inputs=encoder_input, outputs=encoded)

# Define the decoder
decoder_input = keras.Input(shape=(encoding_dim,))
decoded = keras.layers.RepeatVector(window_size)(decoder_input)
decoded = keras.layers.LSTM(encoding_dim, activation='tanh', return_sequences=True)(decoded)
decoded = keras.layers.TimeDistributed(keras.layers.Dense(input_dim))(decoded)
decoder_model = keras.Model(inputs=decoder_input, outputs=decoded)

# Combine the encoder and decoder to create the autoencoder
autoencoder_input = keras.Input(shape=(window_size, input_dim))
encoded = encoder_model(autoencoder_input)
decoded = decoder_model(encoded)
autoencoder = keras.Model(inputs=autoencoder_input, outputs=decoded)

autoencoder.compile(optimizer='adam', loss='mean_squared_error')
autoencoder.fit(X_train, X_train, epochs=19, batch_size=32, shuffle=True, validation_data=(X_test, X_test))
feature_selection_model = encoder_model
#feature_selection_model = keras.Model(inputs=encoded.inputs, outputs=encoded.outputs)
X_encoded = feature_selection_model.predict(sequences)
```
